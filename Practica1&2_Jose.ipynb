**Ejercicio 2**

```python
def process_text(text="Este es un texto de ejemplo", words_to_replace=["ejemplo", "texto"]):
    """
    Procesa una cadena de texto, reemplazando palabras especificadas por asteriscos.

    Args:
        text (str, optional): Cadena de texto a procesar. Defaults to "Este es un texto de ejemplo".
        words_to_replace (list, optional): Lista de palabras a reemplazar. Defaults to ["ejemplo", "texto"].

    Returns:
        tuple: Tupla con el texto procesado y el número de palabras reemplazadas.
    """

    text = text.lower().strip()  # Convertir a minúsculas y eliminar espacios
    count = 0
    for word in words_to_replace:
        text = text.replace(word, "*" * len(word))  # Reemplazar palabra por asteriscos
        count += text.count("*") // len(word)  # Contar palabras reemplazadas
    return text, count

def main():
    # Ejemplo de uso con valores por defecto
    processed_text, replaced_words = process_text()
    print(f"Texto procesado: {processed_text}")
    print(f"Palabras reemplazadas: {replaced_words}")

    # Ejemplo de uso personalizado
    custom_text = "Python es un lenguaje de programación fantástico"
    custom_words = ["python", "lenguaje"]
    result = process_text(custom_text, custom_words)
    print(f"Texto procesado: {result[0]}")
    print(f"Palabras reemplazadas: {result[1]}")

if __name__ == "__main__":
    main()
```

**Ejercicio 3**

```python
def is_prime(n):
    """
    Comprueba si un número n es primo.

    Args:
    n (int): El número a comprobar.

    Returns:
    bool: True si n es primo, False en caso contrario.
    """
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def prime_list(limit):
    """
    Genera una lista de números primos hasta un límite.

    Args:
    limit (int): El límite superior para la generación de números primos.

    Returns:
    list: Lista de números primos hasta el límite.
    """
    return [n for n in range(2, limit + 1) if is_prime(n)]

def check_palindrome(primes):
    """
    Busca palíndromos en una lista de números primos.

    Args:
    primes (list): Lista de números primos.

    Returns:
    list: Lista de palíndromos encontrados en la lista de primos.
    """
    return [p for p in primes if str(p) == str(p)[::-1]]

def categorize_prime(prime):
    """
    Clasifica un número primo en "pequeño", "mediano" o "grande".

    Args:
    prime (int): El número primo a clasificar.

    Returns:
    str: La categoría del número primo.
    """
    if prime < 10:
        return "pequeño"
    elif 10 <= prime < 100:
        return "mediano"
    else:
        return "grande"

def main():
    # Ejemplo de uso
    limit = int(input("Introduce el límite para generar números primos: "))

    primes = prime_list(limit)
    print(f"Números primos hasta {limit}: {primes}")

    palindromes = check_palindrome(primes)
    print(f"Palíndromos en la lista de primos: {palindromes}")

    for prime in primes:
        category = categorize_prime(prime)
        print(f"El número {prime} es {category}.")

if __name__ == "__main__":
    main()
```

**Ejercicio 4**

```python
def leer_numeros():
    """Lee una lista de números enteros del usuario y la devuelve."""
    numeros = []
    while True:
        entrada = input("Ingrese un número (o 'q' para salir): ")
        if entrada == 'q':
            break
        try:
            numero = int(entrada)
            numeros.append(numero)
        except ValueError:
            print("Entrada inválida. Por favor, ingrese un número entero.")
    return numeros


def calcular_suma(numeros):
    """Calcula la suma de todos los números."""
    return sum(numeros)

def calcular_sumas_por_signo(numeros):
    """Calcula la suma de los números positivos y negativos."""
    positivos = sum(num for num in numeros if num > 0)
    negativos = sum(num for num in numeros if num < 0)
    return positivos, negativos

def encontrar_max_min(numeros):
    """Encuentra el número máximo y mínimo de la lista."""
    return max(numeros), min(numeros)

def son_unicos(numeros):
    """Comprueba si todos los números son únicos."""
    return len(set(numeros)) == len(numeros)

def comparar_signos(numeros):
    """Compara la cantidad de números positivos y negativos."""
    positivos, negativos = calcular_sumas_por_signo(numeros)
    return "Hay más números positivos" if positivos > negativos else "Hay más números negativos" if negativos > positivos else "Hay igual cantidad de positivos y negativos"

def menu():
    """Muestra el menú de opciones y devuelve la opción seleccionada."""
    print("Menú:")
    print("1. Calcular la suma total")
    print("2. Calcular sumas por signo")
    print("3. Encontrar máximo y mínimo")
    print("4. Comprobar si los números son únicos")
    print("5. Comparar cantidad de positivos y negativos")
    print("0. Salir")
    opcion = input("Ingrese una opción: ")
    return opcion

def main():
    numeros = leer_numeros()
    while True:
        opcion = menu()
        if opcion == '1':
            print(f"La suma total es: {calcular_suma(numeros)}")
        elif opcion == '2':
            positivos, negativos = calcular_sumas_por_signo(numeros)
            print(f"Suma de positivos: {positivos}")
            print(f"Suma de negativos: {negativos}")
        elif opcion == '3':
            maximo, minimo = encontrar_max_min(numeros)
            print(f"Máximo: {maximo}")
            print(f"Mínimo: {minimo}")
        elif opcion == '4':
            if son_unicos(numeros):
                print("Todos los números son únicos.")
            else:
                print("Hay números repetidos.")
        elif opcion == '5':
            print(comparar_signos(numeros))
        elif opcion == '0':
            break
        else:
            print("Opción inválida.")

if __name__ == "__main__":
    main()
```

Ejercicio 5

```python
tareas = []  # Lista global de tareas


MAX_TAREAS = 5  # Límite máximo de tareas

def add_task(task):
    """Agrega una tarea a la lista global de tareas, si no se ha alcanzado el límite."""
    global tareas
    if len(tareas) < MAX_TAREAS:
        tareas.append(task)
        print(f"Tarea '{task}' agregada.")
    else:
        print(f"Has alcanzado el límite máximo de {MAX_TAREAS} tareas.")

def remove_task(task):
    """Elimina una tarea de la lista global de tareas."""
    global tareas
    if task in tareas:
        tareas.remove(task)
        print(f"Tarea '{task}' eliminada.")
    else:
        print(f"Tarea '{task}' no encontrada.")

def list_tasks():
    """Enumera todas las tareas."""
    print("Lista de tareas:")
    for i, tarea in enumerate(tareas, start=1):
        print(f"{i}. {tarea}")

def task_manager():
    """Administra las operaciones de las tareas."""
    while True:
        print("\nMenú:")
        print("1. Agregar tarea")
        print("2. Eliminar tarea")
        print("3. Listar tareas")
        print("4. Modificar tarea")
        print("0. Salir")
        opcion = input("Ingrese una opción: ")

        if opcion == '1':
            tarea = input("Ingrese la tarea: ")
            add_task(tarea)
        elif opcion == '2':
            tarea = input("Ingrese la tarea a eliminar: ")
            remove_task(tarea)
        elif opcion == '3':
            list_tasks()
        elif opcion == '4':
            def modify_task():
                """Modifica la primera tarea de la lista."""
                global tareas
                if tareas:
                    nueva_tarea = input("Ingrese la nueva tarea: ")
                    tareas[0] = nueva_tarea
                    print(f"Tarea '{tareas[0]}' modificada.")
                else:
                    print("La lista de tareas está vacía.")
            modify_task()
        elif opcion == '0':
            break
        else:
            print("Opción inválida.")

if __name__ == "__main__":
    task_manager()
```

**Ejercicio 6**

```python
def leer_cadenas():
    """Lee una lista de cadenas del usuario y la devuelve."""
    cadenas = []
    while True:
        cadena = input("Ingrese una cadena (o 'q' para salir): ")
        if cadena == 'q':
            break
        cadenas.append(cadena)
    return cadenas

def imprimir_cadenas_con_letra(cadenas, letra):
    """Imprime las cadenas que comienzan con la letra indicada."""
    for cadena in cadenas:
        if cadena.startswith(letra):
            print(cadena)

def contar_subcadenas(cadenas, subcadena):
    """Cuenta cuántas cadenas contienen una subcadena."""
    contador = 0
    for cadena in cadenas:
        if subcadena in cadena:
            contador += 1
    return contador

def encontrar_max_min(cadenas):
    """Encuentra la cadena más larga y la más corta."""
    if cadenas:
        max_cadena = max(cadenas, key=len)
        min_cadena = min(cadenas, key=len)
        return max_cadena, min_cadena
    else:
        return None, None

def comparar_objetos(cadenas, cadena1, cadena2):
    """Compara si dos cadenas son el mismo objeto."""
    if cadena1 in cadenas and cadena2 in cadenas:
        return cadena1 is cadena2
    else:
        return "Cadenas no encontradas en la lista"

def verificar_longitud(cadenas):
    """Muestra las cadenas que tienen una longitud menor a 10 caracteres."""
    cadenas_cortas = []
    for cadena in cadenas:
        if len(cadena) < 10:
            cadenas_cortas.append(cadena)

    if cadenas_cortas:
        print("Cadenas con menos de 10 caracteres:")
        for cadena in cadenas_cortas:
            print(cadena)
    else:
        print("Ninguna cadena tiene menos de 10 caracteres.")

def menu():
    """Muestra el menú de opciones y devuelve la opción seleccionada."""
    print("Menú:")
    print("1. Imprimir cadenas con una letra específica")
    print("2. Contar subcadenas")
    print("3. Encontrar cadena más larga y más corta")
    print("4. Comparar si dos cadenas son el mismo objeto")
    print("5. Verificar longitud de las cadenas")
    print("0. Salir")
    opcion = input("Ingrese una opción: ")
    return opcion

def main():
    cadenas = leer_cadenas()
    while True:
        opcion = menu()
        if opcion == '1':
            letra = input("Ingrese una letra: ")
            imprimir_cadenas_con_letra(cadenas, letra)
        elif opcion == '2':
            subcadena = input("Ingrese una subcadena: ")
            contador = contar_subcadenas(cadenas, subcadena)
            print(f"La subcadena '{subcadena}' aparece {contador} veces.")
        elif opcion == '3':
            max_cadena, min_cadena = encontrar_max_min(cadenas)
            if max_cadena and min_cadena:
                print(f"Cadena más larga: {max_cadena}")
                print(f"Cadena más corta: {min_cadena}")
            else:
                print("La lista de cadenas está vacía.")
        elif opcion == '4':
            cadena1 = input("Ingrese la primera cadena: ")
            cadena2 = input("Ingrese la segunda cadena: ")
            resultado = comparar_objetos(cadenas, cadena1, cadena2)
            print(resultado)
        elif opcion == '5':
            verificar_longitud(cadenas)
        elif opcion == '0':
            break
        else:
            print("Opción inválida.")

if __name__ == "__main__":
    main()
```
